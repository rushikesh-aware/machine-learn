```python
#Which  is most used  programming langauge (use LanguageWorkedWith column)
df['LanguageWorkedWith'].value_counts()
df['LanguageWorkedWith'].value_counts() == df['LanguageWorkedWith'].value_counts().max()
```

```python
#Display how many work remotely on pie chart  [use WorkRemote column]
Less=df[df.WorkRemote=="Less than once per month / Never"].shape[0]
Few=df[df.WorkRemote=="A few days each month"].shape[0]
All=df[df.WorkRemote=="All or almost all the time (I'm full-time remote)"].shape[0]
Z=[Less,Few,All]
Lables=["Less than once per month / Never","A few days each month","All or almost all the time (I'm full-time remote)"]
colors=["violet","white","grey"]
plt.pie(Z,labels=Lables,colors=colors,autopct="%.f%%")
plt.show()
```

```python
#Display how many male and female programmers on pie chart  [use Gender column]
Man=df[df.Gender=="Man"].shape[0]
Woman=df[df.Gender=="Woman"].shape[0]
c=["Male","Female"]
d=[Man,Woman]
import matplotlib.pyplot as plt
colors=["orange","blue"]
plt.pie(d,labels=c,colors=colors,autopct="%.f%%")
plt.show()
```
